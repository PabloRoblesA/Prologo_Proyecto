```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'http://archive.ics.uci.edu/static/public/848/secondary+mushroom+dataset.zip'
r = requests.get(url)

with open('../zip/mushroom_dataset.zip', 'wb') as f:
    f.write(r.content)
```

```python
import zipfile

with zipfile.ZipFile('../zip/mushroom_dataset.zip', 'r') as zip_ref:
    zip_ref.extractall('../zip/mushroom_dataset')

with zipfile.ZipFile('../zip/mushroom_dataset/MushroomDataset.zip', 'r') as zip_ref:
    zip_ref.extractall('../Data/raw/mushroom_class')
```
